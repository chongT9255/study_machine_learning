```python
# 求解 f(x) = x **2 的 f(x) = 2 的x值
# 导入模块
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from matplotlib.pyplot import figimage
```

```python
# 构造目标函数 J(x) = (x ** 2 - 2) ** 2
# 定义目标函数
def J(x):
    return (x ** 2 - 2) ** 2
# 定义梯度函数（导函数）
def gradient(x):
    return 4 * x ** 3 - 8 * x
```

```python
# 用列表保存点的变化轨迹
x_list = []
y_list = []
# 定义超参数学习率 和 x的初始值
alpha = 0.1
x = 1
```

```python
# 重复迭代100次  ==> 当梯度趋近于 0 的时候终止
while np.abs(grad := gradient(x)) > 1e-10:
    y = J(x)
    # 记录当前点的坐标
    x_list.append(x)
    y_list.append(y)
    print(f"x:{x}, J(x):{y}")
    # 计算梯度
    # grad = gradient(x)
    x = x - alpha * grad
print(len(x_list))
```

```python
# 画图
# x = np.arange(-1, 1, 0.01)
x = np.arange(0.9, 1.6, 0.01)
plt.plot(x, J(x), label="f(x)")
plt.plot(x_list, y_list,'r')
plt.scatter(x_list, y_list, c='r')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
