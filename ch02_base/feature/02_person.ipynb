```python
import pandas as pd
```

```python
# 读取数据
advertising = pd.read_csv('../../data/advertising.csv')
print(advertising.head()) # 查看头5条数据
print(advertising.describe()) # 查看相关信息，条数、均值、最小值...
print(advertising.shape)
```

```python
# 数据预处理
# 去掉第一列  ==> id无意义
advertising.drop(advertising.columns[0],axis=1,inplace=True)
# 去掉空值
advertising.dropna(inplace=True) # nplace=True 原地修改
# 提取特征和标签
X = advertising.drop("Sales",axis=1)  # axis=1 指的是整列
y = advertising["Sales"]
print(X.shape)
print(y.shape)
```

```python
person_r = X.corrwith(y,method="pearson")
print(person_r)
```

```python
# 计算相关矩阵
corr_matrix = advertising.corr(method="pearson") # 计算矩阵内的相关系数
print(corr_matrix)
```

```python
# 将相关系数矩阵画成热力图
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
