```python
import numpy as np
```

```python
# 构造特征
a = np.random.randn(100) # 随机生成标准正态分布的数据 均值为 0 方差为 1
print(a)
print(np.var(a))
```

```python
# b = np.random.randn(100)  * 0.1 # 均值为 0 方差为 0.01
b = np.random.normal(5,0.1,size=100)
print(b)
print(np.var(b))
```

```python
# 构造特征向量
X = np.vstack((a,b)).T
print(X)
print(X.shape)
```

```python
# 低方差过滤
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(0.1)
X_filtered = vt.fit_transform(X)
print(X_filtered)
print(X_filtered.shape)
```
