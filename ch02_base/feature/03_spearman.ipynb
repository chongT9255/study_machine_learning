```python
import pandas as pd
from nltk import spearman_correlation
```

```python
# 每周学习时长
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]

# 创建DataFrame对象 Series对象
X = pd.DataFrame(X)
y = pd.Series(y)
print(X.head())
print(y.head())
print(X.shape)
print(y.shape)
```

```python
# 计算斯皮尔曼系数
spearman_corr = X.corrwith(y, method='spearman')
print(spearman_corr)
```
