```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# 生成数据
X = np.random.randn(1000,3)
print(X.shape)
```

```python
# 适用pca进行降维，将3维降到2维
pca = PCA(n_components=2)
pca_x = pca.fit_transform(X)
print(pca_x.shape)
```

```python
# 可视化
# 转换前3维数据可视化
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:,0], X[:,1], X[:,2], c="g")# 第一列、第二列、第三列全部取出来
ax1.set_title("Before PCA(3D)")
ax1.set_xlabel("feature 1")
ax1.set_ylabel("feature 2")
ax1.set_zlabel("feature 2")
 # 转换后2维数据可视化
ax2 = fig.add_subplot(122)
ax2.scatter(pca_x[:,0], pca_x[:,1], c="g")
ax2.set_title("After PCA(2D)")
ax2.set_xlabel("feature 1")
ax2.set_ylabel("feature 2")
plt.show()
```

```python
# 手动构建线性相关的三维特征数据
n = 1000
# 定义两个主成分的方向向量
pc1 = np.random.normal(0,1,n)
pc2 = np.random.normal(0,0.2,n)
# 定义不重要的第三主成分 噪声
noise = np.random.normal(0,0.05,n)
# 构建三维的数据 三个特征的输入数据
X = np.vstack((pc1 + pc2 , pc1 - pc2 , pc2 + noise)).T
print(X.shape)
```

```python
# 适用pca进行降维，将3维降到2维
pca = PCA(n_components=2)
pca_x = pca.fit_transform(X)
print(pca_x.shape)
```

```python
# 可视化
# 转换前3维数据可视化
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:,0], X[:,1], X[:,2], c="g")# 第一列、第二列、第三列全部取出来
ax1.set_title("Before PCA(3D)")
ax1.set_xlabel("feature 1")
ax1.set_ylabel("feature 2")
ax1.set_zlabel("feature 2")
 # 转换后2维数据可视化
ax2 = fig.add_subplot(122)
ax2.scatter(pca_x[:,0], pca_x[:,1], c="g")
ax2.set_title("After PCA(2D)")
ax2.set_xlabel("feature 1")
ax2.set_ylabel("feature 2")
plt.show()
```
