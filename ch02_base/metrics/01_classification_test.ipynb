```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
labels = ['猫','狗']
y_true = ["猫", "猫", "猫", "猫", "猫", "猫", "狗", "狗", "狗", "狗"] # 真实值
y_pred1 = ["猫", "猫", "狗", "猫", "猫", "猫", "猫", "猫", "狗", "狗"] # 预测值
```

```python
matrix_result = confusion_matrix(y_true, y_pred1,labels=labels)
print(pd.DataFrame(matrix_result,index=labels,columns=labels))
```

```python
sns.heatmap(matrix_result,annot=True,fmt="d",cmap="Greens")
```

```python
# 准确率
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true, y_pred1)
print(accuracy)
```

```python
# 精确率
from sklearn.metrics import precision_score
precision = precision_score(y_true, y_pred1,pos_label='猫')
print(precision)
```

```python
# 召回率
from sklearn.metrics import recall_score
recall = recall_score(y_true, y_pred1,pos_label='猫')
print(recall)
```

```python
# F1分数
from sklearn.metrics import f1_score
f1 = f1_score(y_true, y_pred1,pos_label='猫')
print(f1)
```

```python
# 分类任务的评估报告
from sklearn.metrics import classification_report
report = classification_report(y_true,y_pred1,labels=labels,target_names=None)
print(report)
```
