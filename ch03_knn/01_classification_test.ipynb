```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python

X = [[2, 1], [3, 1], [1, 4], [2, 6]] # 特征
y = [0, 1, 0, 1] # 标签

X = np.array(X)
y = np.array(y)
```

```python
# 定义模型
knn = KNeighborsClassifier(n_neighbors=2,weights='distance')
knn.fit(X, y)

x = np.array([[4, 9]])
x_class = knn.predict(x)
print(x_class)
```

```python
# 画图
fig,ax = plt.subplots()
ax.axis('equal')
# 使用布尔索引将两类点分开
X1 = X[y == 0]
X2 = X[y == 1]
# 定义不同颜色
colors = ['r', 'b']
# X1[:, 0] 表示X1的列0 X1[:, 1] 表示X1的列1 ==》 画出X1的列0和列1 取所有行
plt.scatter(X1[:, 0], X1[:, 1], c=colors[0])
plt.scatter(X2[:, 0], X2[:, 1], c=colors[1])
# 新点颜色
x_color = colors[0] if x_class == 0 else colors[1]
plt.scatter(x[:, 0], x[:, 1], c=x_color)
plt.show()
```
