```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# 定义数据
X = [[2, 1], [3, 1], [1, 4], [2, 6]] # 特征
y = [12,33, 12, 10] # 标签
```

```python
# 创建KNN回归模型 weights='distance' 权重(使用欧式距离)
knn = KNeighborsRegressor(n_neighbors=2,weights='distance')
# 训练模型
knn.fit(X, y)

x = [[4,9]]
x_pred = knn.predict(x)
print(x_pred)
```
